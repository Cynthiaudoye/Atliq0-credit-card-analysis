```python
# base on the relation we can see on the chat we need to create another column called credit range
bin_ranges = [300, 450, 500, 550, 600, 650, 700, 750, 800]
bin_labels = [f"{start}-{end-1}" for start, end in zip(bin_ranges, bin_ranges[1:])]
df_credit_clean['credit_score_range'] = pd.cut(df_credit_clean['credit_score'], bins=bin_ranges, labels=bin_labels, include_lowest=True, right=False)
df_credit_clean.head()
```

```python
# getting the mode values base on credit score range and converting to a new dataframe
mode_df = df_credit_clean.groupby('credit_score_range')['credit_limit'].agg(lambda x: x.mode().iloc[0]).reset_index()
mode_df
```

```python
# mergeing the 2 dataframe to hame all in one table
df_credit_clean2 = pd.merge(df_credit_clean, mode_df, on="credit_score_range", suffixes=("", "_mode"))
df_credit_clean2
```

```python
# getting the null values in the credit limit column
df_credit_clean2[df_credit_clean2.credit_limit.isnull()].sample(3)
```

```python
# replacing the null values by filling them up with values from credit limit mode
df_credit_clean3 = df_credit_clean2.copy()
df_credit_clean3['credit_limit'].fillna(df_credit_clean2['credit_limit_mode'], inplace=True)
df_credit_clean3
```

## using IQR to check and remove the outliers in the transactions table on the tran_amount column after fixing the tran_amount that can not be zero

```python
Q1, Q3 = df_trans.tran_amount.quantile([0.25, 0.75])
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower = Q1 - 2*IQR
upper = Q3 + 2*IQR
lower, upper
```

```python
df_trans.describe()
```

```python
df_trans[df_trans.tran_amount<lower]
```

```python
df_trans[df_trans.tran_amount>upper]
```

```python
df_trans_outliers = df_trans[(df_trans.tran_amount<lower)|(df_trans.tran_amount>upper)]
df_trans_outliers
```

```python
trans_mean_per_category = df_trans_normal.groupby('product_category')['tran_amount'].mean()
trans_mean_per_category
```

```python
df_trans.loc[df_trans_outliers.index, 'tran_amount'] = df_trans_outliers['product_category'].map(trans_mean_per_category)
```

```python
# List of categorical columns
cat_cols = ['payment_type', 'platform', 'product_category', 'marital_status', 'age_group']

num_rows = 3

# create subplots
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 5 * num_rows))

# flatten the axes array to make it easier to iterate
axes = axes.flatten()

# create subplots for each categorical column
for i, cat_col in enumerate(cat_cols):
    # calculate the average annual income for each category
    avg_tran_amount_by_category = df_merged_2.groupby(cat_col)['tran_amount'].mean().reset_index()
    
    # Sort the data by 'annual_income' before plotting
    sorted_data = avg_tran_amount_by_category.sort_values(by='tran_amount', ascending=False)
    
    sns.barplot(x=cat_col, y='tran_amount', data=sorted_data, ci=None, ax=axes[i], palette='tab10')
    axes[i].set_title(f'Average transaction amount by {cat_col}')
    axes[i].set_xlabel(cat_col)
    axes[i].set_ylabel('Average transaction amount')
    
    # rotate x-axis labels for better readibility
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)
    
    # Hide any unused subplots
for i in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[i])    
plt.tight_layout()

plt.show()


        
```
